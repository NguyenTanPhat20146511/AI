### Pylot API vs Object-Oriented (OO) API
* Pylot API -> Quickly
* OO API -> Advanced

### Pylot API

```python
x = np.linspace(0,10,1001) #Tạo một mảng bắt đầu từ 0 và kết thức tại 10
# với khoảng chia 1000
plt.plot(x, np.sin(x), color ="blue", linestyle = 'dashed', label = 'sin x'); #Vẽ đồ thị sinx theo x ở trên
plt.plot(x, np.cos(x), color ="red", linestyle = 'dotted', label ='cos x');
plt.title("Sin(x) and Cos(x)")
# Đặt tên trục x và y
plt.xlabel("x") ;
plt.ylabel('sinx');
#Giới hạn đồ thị ở trục x và y
#plt.xlim([0,4]) 
#plt.ylim([-0.9,0.9])
#plt.axis([0,4,-0.7,0.6]) #Có thể sử dụng này để giới hạn đồ thị
#plt.axis("tight") # Hiển thị toàn bộ tổng quan về đồ thị
plt.axis("equal") # Làm cho trục x và trục y bằng nhau
plt.legend() #Sau khi khai báo nhũng label ở trên nhớ xài lệnh này để 
```

### Object-Oriented (OO) API

```python
#1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]
#2. Setup plot
fig, ax = plt.subplots(figsize = (5,5)) #Figure Size = Width & Height of the plot
#fig là hình tổng, còn ax là đồ thị nhỏ
# Chúng ta sử dụng ax để vẽ đồ thị trong đó
#3. Plot the Data
ax.plot(x,y,label ='y = 11x')
ax.set(title = 'A simple Plot', xlabel = 'x axis', ylabel ='y axis');
ax.legend()
```

```python
## OO API
rng = np.random.RandomState(0) #Đối số = 0 dùng để giữ giá trị ngẫu nhiên
#không cho bị thay đổi
x = rng.randn(100) #Lấy một 100 giá trị ngẫu nhiên từ phân phối chuẩn
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)
fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s = sizes, c = colors, cmap ='viridis', alpha = 0.3);
#Bấm Shift + Tab để biết thêm chi tiết
fig.colorbar(img1); #Tạo ra thanh màu thể hiênhj độ Scale màu
```

### 3. Bar
* Vertical (Trục đứng)
* Horizontal (Trục ngang)

```python
#Prepare Data from a fython dictionary
sorf_drink_prices = {"Coca":10,"Pepssi":12,"Sprite":8}
fig, ax = plt.subplots()
ax.bar(sorf_drink_prices.keys(), sorf_drink_prices.values())
#key: lấy từ khoá, values lấy giá trị
ax.set(title="Bach Hoa xanh 's sorf_drink_prices",ylabel ='Prices ($)');
```

```python
fig, ax = plt.subplots()
#Vẽ đồ thị trục ngang dùng lệnh barh (bar + horizontal)
ax.barh(list(sorf_drink_prices.keys()), list(sorf_drink_prices.values())) 
# Nhớ thêm list để chuyển đỏi kiểu dữ liệu
```

```python
students_height[:10], len(students_height)
```

### Subplots

```python
# Option 1:
# Vẽ 4 đồ thị 2 hàng, 2 cột
fig ,((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5)) 
# ax1, ax2 nằm trên cùng 1 hagf và tương tự
#Plot data to each axis
ax1.plot(x,x/2)
ax1.plot(x,x*2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.barh(list(sorf_drink_prices.keys()), list(sorf_drink_prices.values())) 
ax4.hist(students_height, bins = 30); 
```

```python
# Option 2:
# Vẽ 4 đồ thị 2 hàng, 2 cột
fig ,ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))
ax[0,0].plot(x,x/2)
ax[0,0].plot(x,x*2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].barh(list(sorf_drink_prices.keys()), list(sorf_drink_prices.values())) 
ax[1,1].hist(students_height, bins = 30); 
```

### Project: Data Visualization - California Cities


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('data.csv')
cities.head()

cities["area_total_km2"].sort_values()
```

```python
# Extract: Latd (Vĩ độ) and Longd (Kinh độ)
lat, lon = cities['latd'], cities['longd']

population , area = cities['population_total'], cities['area_total_km2']
#Plot using Pyplot API
plt.style.use('seaborn')
plt.scatter(lon,lat, c = np.log10(population), cmap = 'viridis', s = area, linewidths = 3, alpha = 0.5);
plt.axis('equal'); 
plt.xlabel('Longlatitude')
plt.ylabel('Latitude')
plt.colorbar(label = 'log$_{10}$(population)')
plt.clim(3,7) #Cắt ngắn thanh bar từ 3 - 7 


#Create a legend for Cities'Sizes
area_range = [50, 100, 300, 500]  
for area in area_range:
    plt.scatter([],[], s = area, label = str(area)+ 'km$^2$', c= 'k', alpha =0.5)
plt.legend(labelspacing = 2, scatterpoints = 1, title = 'City Area')

plt.title('California Cities: Population and Area Distributions')
```
