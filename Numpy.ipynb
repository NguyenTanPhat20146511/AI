# Array Concatenation and Solitting(Nối các array dài hơn và cắt các array ngắn hơn)

### Concatenation

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
```

```python
np.concatenate((x,y))#axis = 0 (mặc định không cần ghi) thì ta nối theo hàng
```

```python
grid = np.array([[1,2,3],[4,5,6]])
```

```python
grid
```

```python
np.concatenate((grid,grid))
```

```python
np.concatenate((grid,grid), axis = 1) #axis = 1 thì ta nối theo cột
```

```python
#vstack --> chỉ định nối theo hàng
```

```python
x = np.array([1,2,3])
grid = np.array([[9,8,7],[6,5,4]])
```

```python
np.vstack((x, grid))
```

```python
#hstack -->chỉ định nối theo cột
```

```python
y = np.array([[99],[99]])
```

```python
np.hstack((y,grid))
```

# Standar Deviation an Variance (Đô lệch chuẩn và phương sai)

```python
#Example 
dog_height = [600,470,170,430,300]
dog_height = np.array(dog_height)
```

```python
dog_height
```

```python
np.std(dog_height) #Tính độ lệch chuẩn
```

```python
np.var(dog_height) #Tính phương sai
```

```python
np.sqrt(np.var(dog_height)) #Tính lại độ lệch chuẩn
```

```python
weekly_sales
```

```python
# Create a price array
```

```python
butter_prices
```
