# loc and iloc

```python
df.loc[df.Year == 1924] #Lọc những hàng có Year = 1924
```

```python
# Hàm loc có tối đa 2 điều kiện lọc
df.loc[(df.Year == 1924)|(df.Sport =="Skiing")] #Sử dụng điều kiện or lọc 1 trong 2 cái
```

```python
df.loc[(df.Year == 1924)&(df.Sport =="Skiing")] #Sử dụng điều kiện and lọc 2 cái
```

```python
df.loc[(df.Year == 1924)&(df.Sport =="Skiing"),['Year','City','NOC']] #Sử dụng điều kiện and lọc 2 cáiChỉ muốn xem một số cột nhất định
```

```python
df.loc[(df.Year > 2000)&(df.Sport =="Skiing"),['Year','City','NOC']]
```

# Groupby

```python
df.groupby("Sport").apply(print) #Quan sát thây lệnh đã Group lại tất cả các Sport có cùng tên
```

```python
c = df.groupby("Sport")["Year"].sum() 
print(c)
```

```python
c.sort_values() #Sắp xếp theo các thứ tự tăng dần
```

```python
c.sort_values(ascending = False) #Sắp xếp theo các thứ tự tăng dần
```
